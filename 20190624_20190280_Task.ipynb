### NFA Circuit:

**In our NFA graph https://drive.google.com/file/d/19ZfJeTOFzFNZgvv6QqkZ_fTTE656iM_K/view?usp=sharing, and our NFA Table https://drive.google.com/file/d/1O7YakrusNwlczmDo9v_urpDhA_Y6ESpr/view?usp=sharing we try to find "11011" in any string so for our first node if it get '1' it will go to the next step, if it was '0' the first node will loop on it self to begain fro the start.

The next node if it get '1' it will go to the next state, if '0' will go to the first node to break the sequance.

The third node if get '1' it will loop on it self, if '0' it will go to the first node, the fourth node or the second node to solve the overlabe.

The fourth node if get '1' it will go to the fifth node or the second one to solve the overlabe, if get 0 it will go to the first node to break the sequance or to solve the overlabe.

The fifth one if get '1' it will go to the final state, if get '0' it will get back to the fourth one.**

```python
# implement of NFA
NFA_circuit = {
    "q0":{"0":["q0"],"1":["q1"]},
    "q1":{"0":["q0"],"1":["q1","q2"]},
    "q2":{"0":["q0","q1","q3"],"1":["q2"]},
    "q3":{"0":["q0"],"1":["q1","q4"]},
    "q4":{"0":["q3"],"1":["q5"]},
    "q5":{}
     #q0 is the start state
    #q6 is the final state
}
```

```python
def remove_nodes(test_list, item):
  #funnction to remove the doublecated nodes 
    result = [i for i in test_list if i != item]
 
    return result
```

```python
def NFA_occurrences(input,NFA_circuit):
      # This function take the input from the user and go throw the graph to 
      #cuclue the number of occurrences of (11011) in this input

    occurrences = 0
    states = list(NFA_circuit.keys())

    #Initial State
    current_states = [states[0]]

    #This for loop get elments one by one from the input
    for symbol in input:

        next_states = []
        
        #This loop go from this stat to the next
        for state in current_states:
       
            if symbol == '0':
                next_states.append(NFA_circuit[state]["0"])
            elif symbol == '1':
                next_states.append(NFA_circuit[state]["1"])

        next_states = [item for sublist in next_states for item in sublist]

        # copy the next states to make it the current states
        current_states = next_states.copy()

        # if we get the final state then we add 1 to occurrences
        if "q5" in current_states:
            occurrences += 1

            # remove the final state from the current state
            current_states=remove_nodes(current_states, 'q5')

    return occurrences
```

```python
print("Test1 ")
print("#Count = ",NFA_occurrences('110111',NFA_circuit))
print("***********************")
print("Test2")
print("#Count = ",NFA_occurrences('11111',NFA_circuit))
print("***********************")
print("Test3  ")
print("#Count = ",NFA_occurrences('11011011011',NFA_circuit))
print("***********************")
print("Test4 ")
print("#Count = ",NFA_occurrences('1110111011',NFA_circuit))
print("***********************")
```
